```python

import pandas as pd
import matplotlib.pyplot as plt
import statistics as  mode
import numpy as np
import seaborn as sns
import math as sqrt

dat= pd.read_csv(r'C:\Users\HP 850 I7\Desktop\Kaggle_canser_du_sein\Breast_cancer_data.csv',sep=',')

# Features selection to predict breast cancer
df=dat[['mean_radius', 'mean_texture', 'mean_perimeter',
       'mean_smoothness', 'diagnosis']]


# define fonction to scaling the outliers
def imput_outliers(data, features):
    q1 = np.quantile(data[features], 0.25)
    q3= np.quantile(data[features], 0.75)
    # intervalles inter quartile
    iqr = q3 - q1
    # limite du boxplot
    lower_bound = q1 - 1.5*iqr
    uper_bound = q3 + 1.5*iqr
    # replace values up the uper_bound and lower valuers the lower_bound
    data.loc[data[features] < lower_bound, features ] =lower_bound
    data.loc[data[features]> uper_bound, features]=uper_bound 
    ## select features to scalinf
features=['mean_radius', 'mean_texture', 'mean_perimeter','mean_smoothness']

for col in features:
    imput_outliers(df, col)


##select the features
X = df.drop(columns=['diagnosis'])  # Supprimez la colonne 'Outcome' de X
# selection target feature
y= df['diagnosis']

# train_test_split
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.2, random_state=10)


### Import package to do the the GridSearchCV model et selec best parameters
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression

def split_data(data):
    X = df.drop(columns=['diagnosis'])  # Supprimez la colonne 'Outcome' de X
    y = df['diagnosis']  # Créez y avec la colonne 'Outcome' uniquement
    # Split data into training (80%) and testing (20%) sets
    Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.2, random_state=10)
    return Xtrain, Xtest, Ytrain, Ytest

# Function to train a model with hyperparameters 
def train_model(X_train, y_train):
    parametres = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'max_iter': [100, 200, 300],
    'solver': ['liblinear']
    }
    # Train a or many models with hyperparameter tuning
    model = LogisticRegression()
    grid_parma = GridSearchCV(model, parametres, cv=10)
    grid_parma.fit(X_train, y_train)

    model_with_best_estimator = grid_parma.best_estimator_

    # Return best model
    return model_with_best_estimator



# Function to evaluate the model 
def evaluate_model(model, X_test, y_test):
    # Evaluate the best model
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    y_predict = model.predict(X_test)
    acc = accuracy_score(y_test, y_predict)
    #preci = precision_score(y_test, y_predict)
    #recal = recall_score(y_test, y_predict)
    #f1Score = f1_score(y_test, y_predict)

    metric_evaluation = {
        'Accuracy': acc,

    }


    return metric_evaluation


import pandas as pd
import pickle


# Function to deploy the model 
def deploy_model(model, X_test):
    # Enregistrez le modèle à l'aide de Pickle
    with open("model.pkl", "wb") as model_file:
        pickle.dump(model, model_file)

# Main function
def main():
    # Load data
    data =df

    # Split data
    X_train, X_test, y_train, y_test = split_data(data)
    
    # Train a model with hyperparameters
    best_model = train_model(X_train, y_train)
    # Evaluate the model
    metric_evaluation = evaluate_model(best_model, X_test, y_test)
    print(best_model)
    print(metric_evaluation)  # Affichez les performances du modèle

    # Deploy the model (bonus)
    deploy_model(best_model, X_test)

if __name__ == "__main__":
    main()

```
